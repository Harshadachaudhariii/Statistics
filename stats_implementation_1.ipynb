```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings("ignore")
```

```python
# normal distribution/ gaussian distribution >> majority of data point is clustered around mean
mu, sigma = 0, 0.1
s= np.random.normal(mu, sigma, size=10000)
print("count of dps : ",len(s))
s
```

```python
plt.hist(s,40, density=True)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform dist>> all the outcomes has equal prob to occur in range 
u= np.random.uniform(1,0,1000)
plt.hist(u)
```

```python
# log normal distribution >>whose logarithm is normally distributed >>right skewed
l= np.random.lognormal(mean=0 , sigma=0.5, size= 1000)
l = l[np.where(abs(l)< 5)]
plt.hist(l, bins=100)
plt.show()
```

```python
# log normal to normal distribution
sns.distplot(np.log(l))

# transformation --> log, sqrt
sns.distplot(np.sqrt(l))
```

```python
# chi square distribution with dof=1
c = np.random.chisquare(df=1, size=1000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=100)
plt.show()
```

```python
# chi square distribution with dof=3
c = np.random.chisquare(df=3, size=1000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=100)
plt.show()
```

```python
# t distribution with dof =1 when dof is increase then tail of the graph has thicker
t = np.random.standard_t(df=1, size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=100)
plt.show()

# t distribution with dof =4 when dof is increase then tail of the graph has thinner
t = np.random.standard_t(df=4, size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=100)
plt.show()
```

```python
# t distribution with dof =4 when dof is increase then tail of the graph has more thinner and approx. to normal distribution
t = np.random.standard_t(df=31, size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=100)
plt.show()
```

```python
# F distribution 
# dfnum for numerator and dfden for denomintor
f = np.random.f(dfnum=1,dfden=1, size=10000)
f =f [np.where(abs(f)<10)]
plt.hist(f,bins=100)
plt.show()

# dfnum=3 and dfden=10
f = np.random.f(dfnum=3,dfden=10, size=10000)
f =f [np.where(abs(f)<10)]
plt.hist(f,bins=100)
plt.show()
```

*discrete distribution*


```python
# bernoulli distribution >> binomial distribution with 1 trial, n bernoulli distribution trial is binomial distribution 
b = np.random.binomial(n=1, p=0.5, size=10000)
plt.hist(b)
plt.show()
```

```python
# binomial distribution
b = np.random.binomial(n=100,p=0.1, size=1000)
plt.hist(b,bins=20)
plt.show()
```

```python
# poisson distribution
p = np.random.poisson(lam=5, size=1000)
plt.hist(p,bins=20)
plt.show()
```
